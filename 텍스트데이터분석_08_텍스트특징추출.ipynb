# 텍스트 특징 추출 (TF-IDF)

### 1. 샘플 코퍼스로 연습

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.', 
    '자연어처리는 재미있습니다.', 
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.'
]
```

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    return Okt().nouns(text)
```

```python
# TfidfVectorizer 객체 생성
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=my_tokenizer)

# CountVectorizer 객체 생성
# from sklearn.feature_extraction.text import CountVectorizer
# vectorizer = CountVectorizer(tokenizer=my_tokenizer)

# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(sample_corpus)
print(vectorizer.get_feature_names_out())

# 특징 벡터 추출
sample_dtm = vectorizer.transform(sample_corpus)
# type(sample_dtm)
print(sample_dtm.toarray())
```

```python
from konlpy.tag import Okt
def my_movie_tokenizer(text):
    okt = Okt()
    mytags = set(['Noun', 'Adjective'])
    my_stopwords = set('없다 필요없다 하는 한다 의하여 하여 있다 하며 하여야'.split())
    res = [word for word, tag in okt.pos(text) if tag in mytags and word not in my_stopwords and len(word) > 1]
    return res
```

### 2. 다음 영화 리뷰로 연습

```python
# cosine_similarity() 함수를 이용하여 문서 간 유사도 계산
# 전체 단어를 이용하면 차원이 너무 커지므로 max_features 옵션을 주기
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=my_movie_tokenizer, 
                             max_features=100,
                             min_df=2,)

import pandas as pd
datafile = 'data/daum_movie_review.csv'
data_df = pd.read_csv(datafile)

reviews = list(data_df['review'])
vectorizer.fit(reviews)
print(vectorizer.get_feature_names_out())

review_dtm = vectorizer.transform(reviews)
print(review_dtm.shape)
print(review_dtm.toarray())
```
