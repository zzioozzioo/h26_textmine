## BC카드 소비 데이터 분석

#### 1. 데이터 로딩 및 확인

```python
import pandas as pd

bc_data = pd.read_csv('./data/bccard_201906.csv', index_col=0)

bc_data.head(10)
```

```python
bc_data.tail(10)
```

#### 2. 서울시 거주/비거주 고객 비교 분석

#### 2-1. 고객 수

```python
# 서울시 거주
seoul_customers = bc_data[bc_data['CSTMR_MEGA_CTY_NM'] == '서울특별시']
seoul_customers_count = len(seoul_customers)

# 서울시 비거주
non_seoul_customers = bc_data[bc_data['CSTMR_MEGA_CTY_NM'] != '서울특별시']
non_seoul_customers_count = len(non_seoul_customers)

data = [seoul_customers_count, non_seoul_customers_count]
df = pd.DataFrame(data, index=['서울시 거주', '서울시 비거주'], columns=['고객 수'])

df
```

#### 2-2. 총 소비액

```python
# 서울시 거주
seoul_total_amount = seoul_customers['AMT'].sum()

# 서울시 비거주
non_seoul_total_amount = non_seoul_customers['AMT'].sum()

df['총 소비액'] = [seoul_total_amount, non_seoul_total_amount]

df
```

#### 2-3. 카드 이용 건수

```python
# 서울시 거주
seoul_total_cnt = seoul_customers['CNT'].sum()

# 서울시 비거주
non_seoul_total_cnt = non_seoul_customers['CNT'].sum()

df['카드 이용 건수'] = [seoul_total_cnt, non_seoul_total_cnt]

df
```

#### 3. 서울시 편의점 소비 정보 분석

#### 3-1. 서울시 편의점 소비액, 강남구 편의점 소비액

```python
gangpyeon = bc_data[(bc_data['CTY_RGN_NM'] == '강남구') & (bc_data['TP_BUZ_NO'] == 4010)]

pyeon_total = bc_data[bc_data['TP_BUZ_NO'] == 4010]['AMT'].sum()
gn_total = bc_data[(bc_data['CTY_RGN_NM'] == '강남구')]['AMT'].sum()
gangpyeon_total = gangpyeon['AMT'].sum()
gn_live_total = gangpyeon[gangpyeon['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum()
non_gn_live_total = gangpyeon[gangpyeon['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum()

data = [pyeon_total, gn_total, gangpyeon_total, gn_live_total, non_gn_live_total]

df = pd.DataFrame(
    data, 
    index=[
        '편의점', 
        '강남구', 
        '강남구 편의점', 
        '강남구 거주자의 강남구 편의점', 
        '강남구 비거주자의 강남구 편의점'
    ],
    columns=[
        '소비액'
    ]
)

df
```
